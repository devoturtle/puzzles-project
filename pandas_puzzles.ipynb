### **Puzzle 12 — Cleaning Missing Data**
This is one of the most important real world data science skills. Write code that:

departments = pd.DataFrame({\

})

Write code that:

1. Merges using how='left' and prints the result — what happens to Ethan's department?
2. Merges using how='right' and prints the result — what happens to Marketing?
3. Merges using how='inner' and prints the result — who and what get dropped entirely?
4. Merges using how='outer' and prints the result — everyone and everything should appear

After running all four, tell me in your own words what you observe about the differences!

```python
# 1. Ethan's department returns as NaN since there is no value for dept_id 99
# 2. Marketing returns as a NaN row since there is no one assigned to dept_id 40
# 3. Ethan's row got dropped entirely since there was no match between the merging dataframes
# 4. All rows appear


employees = pd.DataFrame({
    "emp_id": [1, 2, 3, 4, 5],
    "name": ["Aiden", "Bella", "Caleb", "Dana", "Ethan"],
    "dept_id": [10, 20, 10, 30, 99]   # 99 doesn't exist in departments
})

departments = pd.DataFrame({
    "dept_id": [10, 20, 30, 40],   # 40 has no employees
    "dept_name": ["Sales", "Engineering", "HR", "Marketing"]
})

df1 = pd.merge(employees, departments, how='left', on='dept_id')
df2 = pd.merge(employees, departments, how='right', on='dept_id')
df3 = pd.merge(employees, departments, how='inner', on='dept_id')
df4 = pd.merge(employees, departments, how='outer', on='dept_id')

print(df1)
print('')
print(df2)
print('')
print(df3)
print('')
print(df4)
```

### **Puzzle 44 — Apply Functions**
This is the next major tool — apply() lets you run custom logic across a column or row, similar to writing your own function and using it everywhere at once.\
students = pd.DataFrame({\
&emsp;"name": ["Aiden", "Bella", "Caleb", "Dana"],\
&emsp;"score": [55, 88, 72, 91]\
})

Write code that:

1. Writes a function called grade(score) that returns "A" for 90+, "B" for 80-89, "C" for 70-79, "D" for 60-69, and "F" below 60
2. Uses .apply() to create a new column called "grade" by applying that function to the score column
3. Prints the final DataFrame

Hint: df["new_column"] = df["existing_column"].apply(function_name) — note you pass the function name without parentheses, since apply calls it for you on each value

```python
students = pd.DataFrame({
    "name": ["Aiden", "Bella", "Caleb", "Dana"],
    "score": [55, 88, 72, 91]
})
print(students)
def grade(score):
    if score >= 90:
        return "A"
    elif score >= 80:
        return "B"
    elif score >= 70:
        return "C"
    elif score >= 60:
        return "D"
    else:
        return "F"
  
students['grade'] = students['score'].apply(grade) 

print(students)
```

### **Puzzle 45 — Apply with Multiple Columns**
This time your function needs two pieces of information, which means applying it across an entire row instead of a single column.\

employees = pd.DataFrame({\
&emsp;"name": ["Aiden", "Bella", "Caleb", "Dana"],\
&emsp;"salary": [55000, 92000, 71000, 110000],\
&emsp;"years": [2, 8, 4, 12]\
})

Write a function called bonus(row) that:

1. If years >= 10, bonus is 15% of salary
2. If years >= 5, bonus is 10% of salary
3. Otherwise, bonus is 5% of salary

Then apply it to create a new "bonus" column.\
Key difference this time: since you need both salary and years together, you apply across the row instead of a column:

df["bonus"] = df.apply(bonus, axis=1)

Inside the function, access values with row["salary"] and row["years"]. Give it a shot — axis=1 is the new concept here!

```python
employees = pd.DataFrame({
    "name": ["Aiden", "Bella", "Caleb", "Dana"],
    "salary": [55000, 92000, 71000, 110000],
    "years": [2, 8, 4, 12]
})

def bonus(row):
  if row['years'] >= 10:
    bonus_amount = .15*row['salary']
    return bonus_amount
  elif row['years'] >= 5:
    bonus_amount = .10*row['salary']
    return bonus_amount
  else:
    bonus_amount = .05*row['salary']
    return bonus_amount

employees['bonus'] = employees.apply(bonus, axis=1) # axis = 1 means apply the function to each row, axis = 0 would apply it to each column
print(employees)
```
