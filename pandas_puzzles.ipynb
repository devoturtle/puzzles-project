### **Puzzle 44 — Apply Functions**
This is the next major tool — apply() lets you run custom logic across a column or row, similar to writing your own function and using it everywhere at once.\
students = pd.DataFrame({\
&emsp;"name": ["Aiden", "Bella", "Caleb", "Dana"],\
&emsp;"score": [55, 88, 72, 91]\
})

Write code that:

1. Writes a function called grade(score) that returns "A" for 90+, "B" for 80-89, "C" for 70-79, "D" for 60-69, and "F" below 60
2. Uses .apply() to create a new column called "grade" by applying that function to the score column
3. Prints the final DataFrame

Hint: df["new_column"] = df["existing_column"].apply(function_name) — note you pass the function name without parentheses, since apply calls it for you on each value

```python
students = pd.DataFrame({
    "name": ["Aiden", "Bella", "Caleb", "Dana"],
    "score": [55, 88, 72, 91]
})
print(students)
def grade(score):
    if score >= 90:
        return "A"
    elif score >= 80:
        return "B"
    elif score >= 70:
        return "C"
    elif score >= 60:
        return "D"
    else:
        return "F"
  
students['grade'] = students['score'].apply(grade) 

print(students)
```

### **Puzzle 45 — Apply with Multiple Columns**
This time your function needs two pieces of information, which means applying it across an entire row instead of a single column.\

employees = pd.DataFrame({\
&emsp;"name": ["Aiden", "Bella", "Caleb", "Dana"],\
&emsp;"salary": [55000, 92000, 71000, 110000],\
&emsp;"years": [2, 8, 4, 12]\
})

Write a function called bonus(row) that:

1. If years >= 10, bonus is 15% of salary
2. If years >= 5, bonus is 10% of salary
3. Otherwise, bonus is 5% of salary

Then apply it to create a new "bonus" column.\
Key difference this time: since you need both salary and years together, you apply across the row instead of a column:

df["bonus"] = df.apply(bonus, axis=1)

Inside the function, access values with row["salary"] and row["years"]. Give it a shot — axis=1 is the new concept here!

```python
employees = pd.DataFrame({
    "name": ["Aiden", "Bella", "Caleb", "Dana"],
    "salary": [55000, 92000, 71000, 110000],
    "years": [2, 8, 4, 12]
})

def bonus(row):
  if row['years'] >= 10:
    bonus_amount = .15*row['salary']
    return bonus_amount
  elif row['years'] >= 5:
    bonus_amount = .10*row['salary']
    return bonus_amount
  else:
    bonus_amount = .05*row['salary']
    return bonus_amount

employees['bonus'] = employees.apply(bonus, axis=1) # axis = 1 means apply the function to each row, axis = 0 would apply it to each column
print(employees)
```

### **Puzzle 48 — Apply with a Helper Function and String Logic**

products = pd.DataFrame({\
&emsp;"product": ["Widget Pro", "Gadget Max", "Tool Basic", "Widget Lite", "Gadget Mini"],\
&emsp;"price": [49.99, 89.99, 19.99, 24.99, 34.99]\
})

Write a function called categorize(row) that:

1. Returns "Premium" if price >= 50
2. Returns "Standard" if price >= 25
3. Returns "Budget" otherwise
4. Also checks if the word "Pro" or "Max" appears in the product name — if so, append " (Flagship)" to the category result

So "Widget Pro" at $49.99 should return "Standard (Flagship)".\
Hint: You can check if a substring exists in a string using the in keyword: "Pro" in row["product"]

```python
products = pd.DataFrame({
    "product": ["Widget Pro", "Gadget Max", "Tool Basic", "Widget Lite", "Gadget Mini"],
    "price": [49.99, 89.99, 19.99, 24.99, 34.99]
})

def categorize(row):
  if row['price'] >=50:
    
    if 'Pro' in row['product'] or 'Max' in row['product']:
      return 'Premium (Flagship)'
    else:
      return 'Premium'
  elif row['price'] >=25:
    
    if 'Pro' in row['product'] or 'Max' in row['product']:
      return 'Standard (Flagship)'
    else:
      return 'Standard'
  else:
    if 'Pro' in row['product'] or 'Max' in row['product']:
      return 'Budget (Flagship)'
    else:
      return 'Budget'

products['category'] = products.apply(categorize, axis=1)
print(products)
```

### **Puzzle 51 — groupby() with .agg()**
You've done simple groupby before — df.groupby("col")["other_col"].mean(). Now let's apply different functions to different columns at once, which is what .agg() unlocks.

sales = pd.DataFrame({\
&emsp;"region": ["East", "East", "West", "West", "East", "West"],\
&emsp;"salesperson": ["Aiden", "Bella", "Caleb", "Dana", "Aiden", "Caleb"],\
&emsp;"revenue": [5000, 7200, 4100, 6800, 5900, 3900],\
&emsp;"units": [50, 72, 41, 68, 59, 39]\
})

Write code that:

1. Groups by region and gets the sum of revenue AND the mean of units, both in a single .agg() call
2. Groups by region and gets multiple stats on revenue alone — sum, mean, max, and count — all at once
3. Groups by region and renames the resulting columns to something cleaner using named aggregation



```python
sales = pd.DataFrame({
    "region": ["East", "East", "West", "West", "East", "West"],
    "salesperson": ["Aiden", "Bella", "Caleb", "Dana", "Aiden", "Caleb"],
    "revenue": [5000, 7200, 4100, 6800, 5900, 3900],
    "units": [50, 72, 41, 68, 59, 39]
})

sales2 = sales.groupby('region').agg(
    total_revenue=('revenue', 'sum'),
    avg_units=('units', 'mean')
)

print(sales2)
print('')

sales3 = sales.groupby('region').agg(
    total_revenue=('revenue', 'sum'),
    avg_revenue=('revenue', lambda x: x.mean().round(2)), #lambda allows for custom aggregation functions, in this case rounding the mean to 2 decimal places. Without the lambda function, the mean would be returned with many decimal places.
    max_revenue=('revenue', 'max'),
    count_revenue=('revenue', 'count')
)
print(sales3)
```

### **Puzzle 53 — Rolling Windows**
Rolling functions calculate something over a sliding window of rows — most commonly used for smoothing out noisy data like daily sales into a rolling average.

daily_sales = pd.DataFrame({\
&emsp;"day": range(1, 11),\
&emsp;"sales": [200, 220, 195, 240, 260, 210, 230, 250, 270, 290]\
})

Write code that:

1. Creates a "rolling_avg_3" column — the average of each day plus the 2 days before it (a 3-day rolling average)
2. Creates a "cumulative_sales" column — the running total of sales up to and including each day
3. Creates a "sales_change" column — the difference between each day's sales and the previous day's sales

```python
daily_sales = pd.DataFrame({
    "day": range(1, 11),
    "sales": [200, 220, 195, 240, 260, 210, 230, 250, 270, 290]
})

daily_sales['three_day_avg'] = daily_sales['sales'].rolling(window=3).mean()
print(daily_sales)
print('')
daily_sales['cumulative_sales'] = daily_sales['sales'].cumsum()
print(daily_sales)
print('')
daily_sales['sales_diff'] = daily_sales['sales'].diff()
print(daily_sales)
```
