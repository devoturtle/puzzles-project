### **Puzzle 30 — seek(), tell(), and truncate()**
Before you code anything, here's what each one does:

tell() — returns the current position of the cursor in the file as a byte count\
seek(n) — moves the cursor to position n — seek(0) rewinds to start\
truncate(n) — cuts the file down to n bytes, removing everything after that point

6. Reads the whole file and prints it
7. Then opens the file in 'r+' mode again, seeks to position 0, and uses truncate() to keep only the first 30 bytes, then reads and prints what's left

```python
with open('data/heroes.txt', 'r+') as file1:
  print(file1.tell())
  print(file1.readline())
  print(file1.tell())
  file1.seek(0)
  print(file1.tell())
  print(file1.read())
  print(file1.tell())

with open('data/heroes.txt', 'r+') as file2:
  file2.seek(0)
  file2.truncate(30)
  print(file2.read())
```

Build a simple log file system that:

logs = [\
&emsp;"INFO: System started",\
&emsp;"WARNING: Low memory",\
&emsp;"ERROR: Connection failed",\
&emsp;"INFO: Retrying connection",\
&emsp;"INFO: Connection restored"\
]

1. Reads and prints only lines that contain the word "ERROR" or "WARNING"
2. Appends a new entry: "INFO: System shutdown"
3. Uses tell() to print the total byte count of the file after appending
4. Opens the file in 'r+' mode, reads all lines into a list, then overwrites the file with only the "INFO" lines — removing all warnings and errors permanently
5. Reads and prints the final cleaned file

```python
logs = [
    "INFO: System started",
    "WARNING: Low memory",
    "ERROR: Connection failed",
    "INFO: Retrying connection",
    "INFO: Connection restored"
]
with open('data/log.txt', 'w') as writefile:
  for line in logs:
    writefile.write(line + '\n')
with open('data/log.txt', 'r') as readfile:
  target_word1 = 'ERROR'
  target_word2 = 'WARNING'
  for line in readfile:
    if target_word1 in line:
      print(line.strip())
    if target_word2 in line:
      print(line.strip())
    else:
      pass

with open('data/log.txt', 'a') as appfile:
  appfile.write('INFO: System shutdown')
  print(appfile.tell())

with open('data/log.txt', 'r+') as cleanfile:
  lines = cleanfile.readlines()
  cleanfile.seek(0)
  cleanfile.truncate() 
  for line in lines:
    if 'INFO' in line:
      cleanfile.write(line)
  cleanfile.seek(0)
  print(cleanfile.read())
```

### **Puzzle 33 — Filtering and Grouping**
Using the same scores.txt file, write code that:

1. Reads and parses the file into a list of tuples like before
2. Separates students into two groups — passing (70 and above) and failing (below 70)
3. Writes passing students to passing.txt and failing students to failing.txt
4. Prints a summary

```python
with open('data/scores.txt', 'r') as file2:
  results = []
  passing = []
  failing = []
  for line in file2:
    parts = line.strip().split(',')
    name = parts[0]
    score = int(parts[1])
    results.append((name, score))
  for (name, score) in results:
    if score <= 69:
        failing.append((name, score))
    else:
        passing.append((name, score))
  pass_count = len(passing)
  print(f'Passing: {pass_count} students.')
  fail_count = len(failing)
  print(f'Failing: {fail_count} students.')
  avg_pass = sum(s for name, s in passing)/len(passing)
  print(f'Average passing grade: {avg_pass:.2f}')
  avg_fail = sum(s for name, s in failing)/len(failing)
  print(f'Average failing grade: {avg_fail:.2f}')
  with open('data/passing.txt', 'w') as passfile:
    for name, score in passing:
      passfile.write(f'{name}, {score}\n')
  with open('data/failing.txt', 'w') as failfile:
    for name, score in failing:
      failfile.write(f'{name}, {score}\n')
```

### **Puzzle 34 — Full Data Pipeline**
This one is the most realistic data task we've done yet. You have a raw messy file to work with:\
raw_data = [\
&emsp;"alice,88,math",\
&emsp;"BOB,72,science",\
&emsp;"Carol,95,math",\
&emsp;"dave,61,science",\
&emsp;"EVE,78,math",\
&emsp;"Frank,84,science",\
&emsp;"GRACE,91,math",\
&emsp;"hank,69,science"\
]\
Write a full pipeline that:

1. Writes this to raw_data.txt
2. Reads it back and cleans each line — names should be properly capitalized, subjects lowercase
3. Parses into a list of tuples (name, score, subject)
4. Writes the cleaned data to clean_data.txt
5. Groups and prints average score per subject:\
&emsp;&emsp;Math average: 88.0\
&emsp;&emsp;Science average: 71.5
6. Prints the top scorer per subject

```python
raw_data = [
    "alice,88,math",
    "BOB,72,science",
    "Carol,95,math",
    "dave,61,science",
    "EVE,78,math",
    "Frank,84,science",
    "GRACE,91,math",
    "hank,69,science"
]

with open('data/raw_data.txt', 'w') as raw_file:
  for line in raw_data:
    raw_file.write(line + '\n')
with open('data/raw_data.txt', 'r') as raw_file:
  clean = []
  for line in raw_file:
    parts = line.strip().split(',')
    parts[0] = parts[0].capitalize()
    parts[1] = int(parts[1])
    parts[2] = parts[2].lower()
    clean.append(parts)

with open('data/clean_data.txt', 'w') as clean_file:
     
  for name, score, subject in clean:
    clean_file.write(f'{name}, {score}, {subject}\n')

with open('data/clean_data.txt', 'r') as clean_file:
  math_class = []
  science_class = []
  for line in clean_file:              
      parts = line.strip().split(',')
      name = parts[0]
      score = int(parts[1])
      subject = parts[2].strip()
      if subject == 'science':
        science_class.append((name, score, subject))
      else:
        math_class.append((name, score, subject))
  print(math_class)
  print(science_class)
  math_avg = sum(score for name, score, subject in math_class)/len(math_class)
  print(f'Math Average: {math_avg:.1f}')
  science_avg = sum(score for name, score, subject in science_class)/len(science_class)
  print(f'Science Average: {science_avg:.1f}')
  max_math = max(math_class, key=lambda student: student[1])
  print(f'{max_math[0]} is the top student in Math with a score of {max_math[1]}')
  max_science = max(science_class, key=lambda student: student[1])
  print(f'{max_science[0]} is the top student in Science with a score of {max_science[1]}')
```

Save all players to a text file in a clean readable format, then write a function that reads that file back and tells you which player has the highest level and which has the lowest HP — without ever referencing the original players list again.

```python
class Player:
    def __init__(self, name, level, hp):
        self.name = name
        self.level = level
        self.hp = hp

players = [
    Player("Aryn", 12, 85),
    Player("Borin", 8, 40),
    Player("Cela", 15, 95),
    Player("Dax", 5, 20)
]

with open('data/challenge3_players.txt', 'w') as file1:
  for line in players:
    
    file1.write(f'{line.name},{line.level},{line.hp}\n')
with open('data/challenge3_players.txt', 'r') as file1:  
  player_list = []
  for line in file1:
    parts = line.strip().split(',')
    name = parts[0]
    level = int(parts[1])
    hp = int(parts[2])
    player_list.append((name, level, hp))
  max_level = max(player_list, key=lambda high:high[1])
  print(f'Highest level player: {max_level[0]} at level {max_level[1]}')
  low_hp = min(player_list, key=lambda low:low[2])
  print(f'Player with lowest hp: {low_hp[0]} with {low_hp[2]} hp')
```
