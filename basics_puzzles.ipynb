### **Puzzle #1 — FizzBuzz (Classic for a reason)**

Write a program that prints the numbers 1 through 30, but:
1. If the number is divisible by 3, print "Fizz" instead
2. If the number is divisible by 5, print "Buzz" instead
3. If it's divisible by both 3 and 5, print "FizzBuzz"
4. Otherwise, just print the number

```python
for num in range(1,31):
    if num % 3 ==0 and num % 5 ==0:
        print('FizzBuzz')
    elif num % 3 == 0:
        print('Fizz')
    elif num % 5 == 0:
        print('Buzz')
    else:
        print(num)
```

### **Puzzle #2 — Temperature Converter**

Write a function that converts Celsius to Fahrenheit. The formula is:
F = (C × 9/5) + 32

Your function should:
1. Accept a Celsius temperature as input
2. Return the Fahrenheit result

Then call the function on these three values and print the results: 0°, 100°, and -40°.

```python
def tempconv(C):
    F = (C*(9/5)) +32
    return F
print(tempconv(0))
print(tempconv(100))
print(tempconv(-40))
```

### **Puzzle #3 — Lists and Loops**
This one introduces lists, which are the stepping stone to working with real datasets.
You have a list of test scores:\
scores = [88, 92, 45, 70, 61, 99, 34, 78, 85, 55]

Write code that:
1. Prints every score that is below 70 (these students are failing)
2. Calculates and prints the average score of the whole list

```python
scores = [88, 92, 45, 70, 61, 99, 34, 78, 85, 55]
for num in scores:
    if num <=69:
        print(f'{num} is a failing grade')
    else:
        pass
def avg_score(scores):
    x = sum(scores)/len(scores)
    return x
print(avg_score(scores))
```

### **Puzzle #4 — Dictionaries**
You have this dictionary of a player's stats in a game:\
player = {
    "name": "Aryn",
    "health": 72,
    "attack": 15,
    "defense": 9,
    "level": 4
}\
Write code that:

1. Prints the player's name and level in the format: Aryn is level 4
2. Increases attack by 5 (the player leveled up)
3. Prints the updated attack value

```python
player = {
    "name": "Aryn",
    "health": 72,
    "attack": 15,
    "defense": 9,
    "level": 4
}
print(f'{player["name"]} is level {player["level"]}')
def lvlup(player):
    player["health"] += 10
    player["attack"] += 5
    player["defense"] += 5
    player["level"] += 1
print(player)
lvlup(player)
print(player)
```

### **Puzzle #5 — Putting it together**
You have a list of players:\
players = [
    {"name": "Aryn", "health": 72, "level": 4},
    {"name": "Borin", "health": 45, "level": 7},
    {"name": "Cela", "health": 91, "level": 2},
    {"name": "Dax", "health": 28, "level": 9}
]\
Write code that:

1. Prints the name and health of any player whose health is below 50 (they need a potion)
2. Finds and prints the name of the highest level player

```python
players = [
    {"name": "Aryn", "health": 72, "level": 4},
    {"name": "Borin", "health": 45, "level": 7},
    {"name": "Cela", "health": 91, "level": 2},
    {"name": "Dax", "health": 28, "level": 9}
]
for player in players:
    if player["health"] < 50:
        print(f'{player["name"]}: {player["health"]}')
        
def best():
   return max(players, key=lambda p: p["level"])
print(f'Best player: {best()["name"]}')
```

### **Puzzle 6A — Dictionary Practice**
Start with this dictionary:\
inventory = {
    "arrows": 30,
    "potions": 4,
    "gold": 150,
    "torches": 12
}\
Write code that:

1. Adds a new item "rope" with a value of 2
2. Removes "torches" entirely
3. Prints each item and quantity in the format: arrows: 30

```python
inventory = {
    "arrows": 30,
    "potions": 4,
    "gold": 150,
    "torches": 12
}
inventory["rope"] = 2
del(inventory["torches"])
for item in inventory:
    print(f'{item}: {inventory[item]}')
```

### **Puzzle 7 — List Comprehensions**
You have this list of numbers:\
numbers = [4, 17, 23, 8, 42, 15, 6, 99, 31, 12]\
Using list comprehensions, write three separate one-liners that create new lists:

1. Every number doubled
2. Only numbers greater than 20
3. The square of every number that is even

```python
numbers = [4, 17, 23, 8, 42, 15, 6, 99, 31, 12]
result = [x*2 for x in numbers]
print(result)
result2 = [x for x in numbers if x > 20]
print(result2)
result3 = [x**2 for x in numbers if x % 2 == 0]
print(result3)
```
