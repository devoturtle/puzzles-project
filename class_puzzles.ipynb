3. Method: show_averages() — loops through all students and prints "[name]: [average]" for each
4. Method: top_student() — finds and prints the name of the student with the highest average

```python
class Student():

  def __init__(self, name):
    self.name = name
    self.grades = []

  def add_grade(self, grade):
    self.grades.append(grade)
    print(f'{grade} added for {self.name}')

  def average(self):
    return sum(self.grades)/len(self.grades)
      

class Classroom():

  def __init__(self, subject):
    self.subject = subject
    self.students = []

  def enroll(self, Student):
    
    self.students.append(Student)
    print(f'{Student.name} enrolled in {self.subject}')

  def show_averages(self):
    for student in self.students:
        if not student.grades:
            print(f'{student.name} has no grades.')
        else:
            print(f'{student.name}: {student.average()}')

  def top_student(self):
    try: 
        best = max(self.students, key=lambda student: student.average())
        print('Top Student:', best.name)
    except ValueError:
        print('No students available.')
    except ZeroDivisionError:
        print('No grades available to determine top student.')
    
s1 = Student("Alice")
s2 = Student("Bob")
s3 = Student("Carol")
room = Classroom("Math")
room.enroll(s1)
room.enroll(s2)
room.enroll(s3)
room.show_averages()
room.top_student()
s1.add_grade(88)
s1.add_grade(92)
s2.add_grade(74)
s2.add_grade(81)
s3.add_grade(95)
s3.add_grade(98)


room.show_averages()
room.top_student()
```

### **Puzzle 21 — Game Tournament**
Three classes:

&emsp;Fighter class:
1. Attributes: name, wins (starts at 0), losses (starts at 0)
2. Method: record_win() — adds 1 to wins
3. Method: record_loss() — adds 1 to losses
4. Method: win_rate() — returns wins divided by total fights, or returns 0 if no fights yet
5. Method: show_record() — prints "[name] - Wins: [wins] Losses: [losses] Win Rate: [win_rate]%"

&emsp;Match class:
1. Attributes: fighter1, fighter2, winner (starts as None)
2. Method: decide_winner() — whoever has the higher win_rate() wins. Call record_win() on the winner and record_loss() on the loser. Set self.winner to the winning fighter. Print "[name] wins the match!"

&emsp;Tournament class:
1. Attributes: name, matches (empty list)
2. Method: add_match(match) — adds a Match object to matches
3. Method: run_tournament() — loops through all matches and calls decide_winner() on each
4. Method: champion() — finds and prints the fighter with the most wins across all matches

```python
class Fighter():

  def __init__(self, name):
    self.name = name
    self.wins = 0
    self.losses = 0

  def record_win(self):
    self.wins += 1
  def record_loss(self):
    self.losses += 1
  def win_rate(self):
    if (self.wins + self.losses) == 0:
      return 0
    else:
      return (self.wins/(self.wins + self.losses))*100
  def show_record(self):
    print(f'{self.name} - Wins: {self.wins} Losses: {self.losses} Win Rate: {self.win_rate():.2f}%')

class Match():

  def __init__(self, fighter1, fighter2):
    self.fighter1 = fighter1
    self.fighter2 = fighter2
    self.winner = None
  def decide_winner(self):
    if self.fighter1.win_rate() > self.fighter2.win_rate():
      self.fighter1.record_win()
      self.fighter2.record_loss()
      self.winner = self.fighter1
      print(f'{self.fighter1.name} wins the match against {self.fighter2.name}!')
    else:
      self.fighter2.record_win()
      self.fighter1.record_loss()
      self.winner = self.fighter2
      print(f'{self.fighter2.name} wins the match against {self.fighter1.name}!')

class Tournament():

  def __init__(self, name):
    self.name = name
    self.matches = []
  def add_match(self, Match):
    self.matches.append(Match)
  def run_tournament(self):
    for fight in self.matches:
      fight.decide_winner()
  
  def champion(self):
    fighters = []
    for match in self.matches:
        if match.fighter1 not in fighters:
            fighters.append(match.fighter1)
        if match.fighter2 not in fighters:
            fighters.append(match.fighter2)
    champ = max(fighters, key=lambda fighter: fighter.wins)
    print(f'Champion is: {champ.name}')

f1 = Fighter("Aryn")
f2 = Fighter("Borin")
f3 = Fighter("Cela")

f1.record_win()
f1.record_win()
f1.record_loss()
f2.record_win()
f2.record_loss()
f2.record_loss()
f3.record_loss()
f3.record_loss()
f3.record_loss()
f1.show_record()
f2.show_record()
f3.show_record()

m1 = Match(f1, f2)
m2 = Match(f2, f3)
m3 = Match(f1, f3)

t = Tournament("Spring Cup")
t.add_match(m1)
t.add_match(m2)
t.add_match(m3)
t.run_tournament()
f1.show_record()
f2.show_record()
f3.show_record()
t.champion()
```

### **Puzzle 23 — Object Passing Practice**
Two simple classes:

&emsp;Dog class:
1. Attributes: name, breed
2. Method: speak() — returns "[name] says Woof!"

&emsp;DogPark class:
1. Attributes: park_name, dogs (empty list)
2. Method: enter(dog) — takes a Dog object, adds it to the list, prints "[name] entered the park"
3. Method: roll_call() — loops through all dogs and calls speak() on each one, printing the result

```python
class Dog():
  def __init__(self, name, breed):
    self.name = name
    self.breed = breed

  def speak(self):
    return f'{self.name} says Woof!'

class DogPark():
  def __init__(self, park_name):
    self.park_name = park_name
    self.dogs = []

  def enter(self, dog):
    self.dogs.append(dog)
    print(f'{dog.name} ({dog.breed}) entered the park.')

  def roll_call(self):
    for doggy in self.dogs:
      print(doggy.speak())

d1 = Dog("Rex", "Labrador")
d2 = Dog("Bella", "Poodle")
d3 = Dog("Max", "Bulldog")

park = DogPark("Sunny Park")
park.enter(d1)
park.enter(d2)
park.enter(d3)
park.roll_call()
```

### **Puzzle 24 — Hospital System**
&emsp;Patient class:
1. Attributes: name, age, condition
2. Method: summary() — returns "[name], Age: [age], Condition: [condition]"

&emsp;Doctor class:
1. Attributes: name, specialty
2. Method: diagnose(patient) — takes a Patient object and prints "Dr. [doctor name] is treating [patient name] for [condition]"

&emsp;Hospital class:
1. Attributes: hospital_name, patients (empty list), doctors (empty list)
2. Method: admit(patient) — adds a Patient object, prints "[name] has been admitted"
3. Method: hire(doctor) — adds a Doctor object, prints "Dr. [name] has joined the staff"
4. Method: show_patients() — loops through patients and prints each one's summary()
5. Method: treat_all(doctor) — takes a Doctor object and calls diagnose() on every patient

```python
class Patient():
  def __init__(self, name, age, condition):
    self.name = name
    self.age = age
    self.condition = condition
  def summary(self):
    return f'{self.name}, Age: {self.age}, Condition: {self.condition}'

class Doctor():
  def __init__(self, name, speciality):
    self.name = name
    self.speciality = speciality
  def diagnose(self, patient):
    print(f'Dr. {self.name} is treating {patient.name} for {patient.condition}.')

class Hospital():
  def __init__(self, hospital_name):
    self.hospital_name = hospital_name
    self.patients = []
    self.doctors = []
  def admit(self, patient):
    self.patients.append(patient)
    print(f'{patient.name} has been admitted.')
  def hire(self, doctor):
    self.doctors.append(doctor)
    print(f'Dr. {doctor.name} has joined the staff.')
  def show_patients(self):
    for sick in self.patients:
      print(sick.summary())
  def treat_all(self, doctor):
    for sick in self.patients:
      doctor.diagnose(sick)

p1 = Patient("Alice", 34, "Flu")
p2 = Patient("Bob", 52, "Broken Arm")
p3 = Patient("Carol", 28, "Migraine")

d1 = Doctor("Smith", "General Practice")
d2 = Doctor("Jones", "Neurology")

h = Hospital("City Hospital")
h.admit(p1)
h.admit(p2)
h.admit(p3)
h.hire(d1)
h.hire(d2)
h.show_patients()
h.treat_all(d1)
h.treat_all(d2)
```
